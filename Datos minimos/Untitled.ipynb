```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

```python
edad_x=[20,25,30,35,40,45,50,55,60,65]
salario_cd_y=[20000,30000,33000,40000,42000,45000,47000,60000,70000,80000]
salario_fed_y=[10000,15000,17000,27000,30000,32000,35000,40000,49000,50000]
salario_bed_y=[15000,20000,25000,27000,35000,40000,45000,50000,55000,60000]
```

```python
plt.plot(edad_x,salario_cd_y)
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y)
plt.title('Salario medio de un Cientifico de Datos en dolares por edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y, label='Cientifico de datos',linewidth=3,marker='.',markersize=20)
plt.plot(edad_x,salario_fed_y, label='Front End',linestyle='--')
plt.plot(edad_x,salario_bed_y, label='Back End')
plt.title('Salario medio de un Cientifico de Datos en dolares por edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

plt.plot(edad_x,salario_cd_y, label='Cientifico de datos',marker='.',markersize=20)
plt.plot(edad_x,salario_fed_y, label='Front End',linestyle='--')
plt.plot(edad_x,salario_bed_y, label='Back End')

salario_medio= np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio,label='Salario Medio',color='#000000',linewidth=3)

plt.title('Salario medio de un Cientifico de Datos en dolares por edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.bar(edad_x,salario_cd_y, label='Cientifico de datos')
plt.bar(edad_x,salario_fed_y, label='Front End')
plt.bar(edad_x,salario_bed_y, label='Back End')

salario_medio= np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.bar(edad_x, salario_medio,label='Salario Medio',color='#000000')

plt.title('Salario medio de un Cientifico de Datos en dolares por edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')
plt.grid(True)
plt.legend()

plt.show()
```

```python
width=0.1

edad_np_x= np.arange(len(edad_x))

plt.bar(edad_np_x -2*width,salario_cd_y, label='Cientifico de datos', width=width)
plt.bar(edad_np_x ,salario_fed_y, label='Front End',width=width)
plt.bar(edad_np_x +2*width,salario_bed_y, label='Back End',width=width)

salario_medio= np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3
plt.bar(edad_np_x +4*width, salario_medio,label='Salario Medio',color='#000000',width=2*width)

plt.title('Salario medio de un Cientifico de Datos en dolares por edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.xticks(ticks=edad_np_x, labels=edad_x)
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.pie(salario_cd_y, labels=edad_x, autopct='%1.1f%%')
plt.title('Edad y Salario')
plt.show()
```
